```python
sns.pairplot(df)
plt.show()
```

```python
sns.boxplot(x = df['Engine HP'])
plt.show()
```

```python
df.Make.value_counts().nlargest(40).plot(kind='bar',figsize=(10,5))
plt.title("Number of cars by making production")
plt.ylabel("No. of cars")
plt.xlabel("Make of Cars")
plt.show()
```

```python
fig,axis = plt.subplots(figsize=(10,6))
axis.scatter(df['Engine Cylinders'],df['MSRP'])
axis.set_xlabel('Engine Cylinders')
axis.set_ylabel('MSRP')
plt.title("Relation between number of cylinders & the price of the car")
plt.show()
```
